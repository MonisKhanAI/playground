```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('data/amazon-reviews/train_40k.csv')
df.head()
```

```python
cat1 = list(df.Cat1.unique()); cat1
```

```python
hier = {}
for cat in cat1:
    sub_cats = df[df['Cat1'] == cat].Cat2.unique()
    hier[cat] = list(sub_cats)
```

```python
hier
```

```python
stats_data = []
for cat in cat1:
    count = sum(df['Cat1'] == cat)
    sub_cats = len(hier[cat])
    stats_data.append({"cat":cat, "nr_sub_cats": sub_cats, "nr_docs": count})
sdf = pd.DataFrame(stats_data)
sdf.head()    
```
