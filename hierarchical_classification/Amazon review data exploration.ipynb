```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('data/train_40k.csv')
df.head()
```

# Category statistics

```python
cat1 = list(df.Cat1.unique()); cat1
```

```python
hier = {}
for cat in cat1:
    sub_cats = df[df['Cat1'] == cat].Cat2.unique()
    hier[cat] = list(sub_cats)
```

```python
hier
```

```python
stats_data = []
tot_count = 0
for cat in cat1:
    count = sum(df['Cat1'] == cat)
    sub_cats = len(hier[cat])
    stats_data.append({"cat":cat, "nr_sub_cats": sub_cats, "nr_docs": count})
    tot_count += count
sdf = pd.DataFrame(stats_data)
sdf.head()   
print(tot_count)
```

```python
df['text_feature'] = df.apply(lambda x: f"{x.Title} {x.Text}", axis=1)
```

```python
df.head()
```

```python
for cat in cat1:
    print(cat)
    column_name = f"labels-{cat.replace(' ', '-')}"
    df[column_name] = df['Cat2'].apply(lambda x: x if x in hier[cat] else 'n/a')
```
